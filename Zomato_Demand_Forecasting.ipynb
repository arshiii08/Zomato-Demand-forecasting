```python
# Import the pandas library with the alias 'pd'
import pandas as pd

# Let's load the newly provided dataset to inspect its structure and ensure it has time-related data for time series forecasting
zomato_dataset_path = '/content/Zomato Dataset.csv'
zomato_ts_data = pd.read_csv(zomato_dataset_path)

# Display the first few rows to understand the structure of the data
zomato_ts_data.head()
```

```python
# Convert the 'Order_Date' column to datetime format
zomato_ts_data['Order_Date'] = pd.to_datetime(zomato_ts_data['Order_Date'], format='%d-%m-%Y')

# Aggregate the data to get the number of orders per day
daily_orders = zomato_ts_data.groupby('Order_Date').size().reset_index(name='Order_Count')

# Display the aggregated data
daily_orders.head()
```

```python
# Now let's build a time series forecasting model using the ARIMA model to predict future demand

from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Set the 'Order_Date' column as the index for time series
daily_orders.set_index('Order_Date', inplace=True)

# Fit an ARIMA model (we'll start with a basic ARIMA(1,1,1) model)
model = ARIMA(daily_orders['Order_Count'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast the next 30 days
forecast = model_fit.forecast(steps=30)

# Plot the historical data and the forecast
plt.figure(figsize=(10, 6))
plt.plot(daily_orders.index, daily_orders['Order_Count'], label='Historical Orders')
plt.plot(pd.date_range(start=daily_orders.index[-1], periods=30, freq='D'), forecast, label='Forecast', color='red')
plt.title('Zomato Order Demand Forecast')
plt.xlabel('Date')
plt.ylabel('Order Count')
plt.legend()
plt.show()
```

```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

# Prepare the data for Prophet model
prophet_data = daily_orders.reset_index()[['Order_Date', 'Order_Count']]
prophet_data.columns = ['ds', 'y']  # Prophet requires 'ds' for date and 'y' for target

# Initialize and fit the Prophet model
prophet_model = Prophet()
prophet_model.fit(prophet_data)

# Forecast the next 30 days
future_dates = prophet_model.make_future_dataframe(periods=30)
forecast_prophet = prophet_model.predict(future_dates)

# Plot the forecast
fig = prophet_model.plot(forecast_prophet, xlabel='Date', ylabel='Order Count')
plt.title('Zomato Order Demand Forecast using Prophet')
plt.show()
```
